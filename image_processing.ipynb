```python
import numpy as np
```

```python
def translate (img,x,y):
  (h,w)=img.shape[:2]
  M=np.float32([[0,1,x],[1,0,y]])
  shifted=cv2.warpAffine(img,M,(w,h))
  return shifted

```

```python
shifted=translate(im,10,50)
```

```python
cv2_imshow(shifted)
```

```python
def rotate (img, angle ,center=None, scale=1):
  (h,w)=img.shape[:2]
  if center is None:
    center=(w/2,h/2)
    M=cv2.getRotationMatrix2D(center,angle, scale)
    rotated=cv2.warpAffine(img,M,(w,h))
    return rotated
```

```python
rotated=rotate(im,62,scale=0.3)
```

```python
cv2_imshow(rotated)
```
